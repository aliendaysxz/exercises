```python
import pandas as pd
import numpy as np
```

```python
potholes = pd.DataFrame({
    'length':[5.1, np.nan, 6.2, 4.3, 6.0, 5.1, 6.5, 4.3, np.nan, np.nan],
    'width':[2.8, 5.8, 6.5, 6.1, 5.8, np.nan, 6.3, 6.1, 5.4, 5.0],
    'depth':[2.6, np.nan, 4.2, 0.8, 2.6, np.nan, 3.9, 4.8, 4.0, np.nan],
    'location':pd.Series(['center', 'north edge', np.nan, 'center', 'north edge', 'center', 'west edge',
                          'west edge', np.nan, np.nan], dtype='string')
})
print(potholes)
```

```python
pd.isna(potholes)
```

```python
potholes['nan_count'] = potholes.apply(lambda x: x.isna().sum(), axis=1)
print(potholes)
```

```python
potholes.drop([potholes.loc[np.where((potholes['nan_count'].to_numpy()) > potholes.shape[1]/2)].index[0]])
```

```python
a = pd.DataFrame(potholes.apply(lambda x: x.isna().sum(), axis=0), columns=['average_nan_per_column'])
print(a)
```

```python
################################
p = potholes.copy()
drop_rows = p.isnull().sum(axis=1) > p.shape[1]/2
print(drop_rows)
```

```python
p.fillna(p.mean(), inplace=True)
print(p.mean(),'\n')
print(p)
```

```python
p.location.fillna(p.location.mode().iat[0], inplace=True)
print(p)
```

```python
p = p.loc[~drop_rows]
print(p)
```
